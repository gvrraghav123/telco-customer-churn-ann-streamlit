```python
import streamlit as st
import pandas as pd
import numpy as np
import pickle
import os

from tensorflow.keras.models import load_model

# -----------------------------
# Page Config
# -----------------------------
st.set_page_config(page_title="Customer Churn Prediction", page_icon="📊")

# -----------------------------
# Safe loader
# -----------------------------
@st.cache_resource
def load_assets():

    model_path = "churn_model.keras"

    if not os.path.exists(model_path):
        st.error(f"Model file not found: {model_path}")
        st.stop()

    model = load_model(model_path)

    scaler = pickle.load(open("scaler.pkl", "rb"))
    encoders = pickle.load(open("encoders.pkl", "rb"))

    return model, scaler, encoders


model, scaler, encoders = load_assets()

# -----------------------------
# UI
# -----------------------------
st.title("Customer Churn Prediction App")

gender = st.selectbox("Gender", ["Male", "Female"])
tenure = st.number_input("Tenure", min_value=0)
monthly_charges = st.number_input("Monthly Charges", min_value=0.0)

# -----------------------------
# Encoding helper
# -----------------------------
def encode(col, value):
    return encoders[col].transform([value])[0]

# -----------------------------
# Prediction
# -----------------------------
if st.button("Predict"):

    gender_enc = encode("Gender", gender)

    input_data = np.array([[gender_enc, tenure, monthly_charges]])
    input_scaled = scaler.transform(input_data)

    prediction = model.predict(input_scaled)[0][0]
    result = 1 if prediction > 0.5 else 0

    if result == 1:
        st.error("Customer will CHURN ❌")
    else:
        st.success("Customer will NOT churn ✅")
```
